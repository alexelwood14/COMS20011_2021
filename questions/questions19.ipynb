```python
import math

import numpy as np
import matplotlib.pyplot as plt
```

## Question 1

Plotting superposition:

```python
n = [1/4, 1, 2]

x = np.linspace(-3, 3, 100)
y1 = np.sin(2*math.pi*n[0]*x)
y2 = np.sin(2*math.pi*n[1]*x)
y3 = np.sin(2*math.pi*n[2]*x)

y4 = y1+y2+y3

fig, ax = plt.subplots()
ax.plot(x, y1, c='blue')
ax.plot(x, y2, c='blue')
ax.plot(x, y3, c='blue')
ax.plot(x, y4, c='red')
```
